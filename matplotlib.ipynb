# GETTING STARTED WITH MATPLOTLIP
### SYNTAX
#### IMPORT MATPLOLIB.pyplot as plt

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6])
y = np.array([0,25])
plt.plot(x,y)
plt.show()
```

## DIFFRENT TYPES OF PLOT

### LINE CHART
##### plt.plot(arrayname, linestye = "stylename")

```python
points = np.array([4,7,8,2,10])
```

```python
plt.plot(points, linestyle = "dashed")
plt.show()
```

```python
a1 = np.array([1,2,3,4,5,6])
plt.plot(a1, linestyle = "dotted")
plt.show()
```

## BAR CHART
##### plt.bar(arrayname, linestye = "stylename")

```python
x = np.array(["MONDAY" , "WEDNESDAY" , "FRIDAY" ])
y = np.array([50,45,64])
plt.bar(x,y)
plt.show()
```

## SCATER PLOT
#### plt.scatter(arrayname, arr)

```python
x1 = np.array([5 , 7 , 8 , 10 , 15])
y1 = np.array([34 , 21 , 6 , 43 , 39])
plt.scatter(x1,y1)
plt.show()
```

## PIE CHART
#### plt.pie(arrayname)

```python
x2 = np.array([5 , 7 , 8 , 10 , 15])
plt.pie(x2)
plt.show()
```

## HISTOGRAM
#### plt.hist(arrayname)

```python
h1 = np.random.normal(170,10,250)
plt.hist(h1)
plt.show()
```
